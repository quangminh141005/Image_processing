```python
image  = cv2.imread('white_dot.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python

image = cv2.imread('white_dot.png', cv2.IMREAD_GRAYSCALE)
dft_img = cv2.dft(np.float32(image), flags=cv2.DFT_COMPLEX_OUTPUT)

# Calculate magnitude
magnitude = cv2.magnitude(dft_img[:, :, 0], dft_img[:, :, 1])

# Apply log scale for better visualization
magnitude_log = 20 * np.log(magnitude + 1)  # +1 to avoid log(0)

plt.imshow(magnitude_log, cmap='gray')
plt.title('Magnitude Spectrum')
plt.axis('off')
plt.show()
```
